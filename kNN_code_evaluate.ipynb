## kNN 

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
```

```python
n = 300 # 300 sampales
centers = [[2,2], [4,2], [2,4]]
cluster_std = [0.5, 0.5, 0.5]

X, y = make_blobs(
    n_samples = n,
    centers = centers,
    cluster_std = cluster_std,
    random_state = 0
)

print(X[:10], y[:10])
```

```python
plt.figure(figsize=(8,6))

colors = ['#FF9999', '#66B2FF','#99FF99']
for i in range(3):
    plt.scatter(
        X[y == i, 0],
        X[y == i, 1],
        c = colors[i],
        label=f'Blob {i+1}',
        alpha = 0.6
    )
    

plt.title('Three 2D Gaussian Blobs', fontsize = 14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42
)

knn = KNeighborsClassifier(n_neighbors=3) #Euclid distance, k =5 Mahhatan 
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

plt.figure(figsize=(8,6))

colors = ['#FF9999', '#66B2FF','#99FF99']
for i in range(3):
    plt.scatter(
        X[y == i, 0],
        X[y == i, 1],
        c = colors[i],
        label=f'Blob {i+1}',
        alpha = 0.6
    )

plt.scatter(
    X_test[:, 0], 
    X_test[:, 1], 
    c='yellow', 
    edgecolors='black',  # Viền đen
    s=100,               # Kích thước lớn hơn
    marker='*',          # Hình dạng sao
    label='Test Points'
)

plt.title('Three 2D Gaussian Blobs', fontsize = 14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print('Classification accurancy', accuracy)
```
